```python
import turtle
scr = turtle.Screen()
scr.bgcolor("black")
turtle = turtle.Turtle()
turtle.shape("arrow")
turtle.color("pink")
turtle.width(5)

def A(size):
    turtle.rt(16)
    turtle.forward(size)
    turtle.rt(150)
    turtle.fd(size)
    turtle.backward(size/2)
    turtle.rt(105)
    turtle.fd(size/3)
    
def D(size):
    turtle.fd(60)
    turtle.rt(90)
    turtle.fd(9)
    for i in range(13):
        turtle.rt(13)
        turtle.fd(7)

def H(size):
    turtle.fd(60)
    turtle.backward(30)
    turtle.rt(90)
    turtle.fd(30)
    turtle.lt(90)
    turtle.fd(30)
    turtle.backward(60)
    
def T(size):
    turtle.fd(size)
    turtle.lt(90)
    turtle.fd(10)
    turtle.backward(20)
    
def P(size):
    turtle.fd(60)
    turtle.rt(90)
    turtle.fd(7)
    for i in range(8):
        turtle.rt(20)
        turtle.fd(5)
        
def Y(size):
    turtle.fd(40)
    turtle.left(60)
    turtle.fd(20)
    turtle.fd(size)
    turtle.backward(20)
    turtle.rt(90)
    turtle.fd(35)
    
def R(size):
    turtle.fd(60)
    turtle.rt(90)
    turtle.fd(7)
    for i in range(15):
        turtle.rt(12)
        turtle.fd(3)
    turtle.lt(120)
    turtle.fd(49)
def M(size):
    turtle.fd(int(size/2))
    turtle.rt(135)
    turtle.fd(int(size/3))
    turtle.lt(90)
    turtle.fd(int(size/3))
    turtle.rt(135)
    turtle.fd(int(size/2))    
    
def E(size):
    turtle.rt(90)
    turtle.fd(int(size/3))
    turtle.back(int(size/3))
    turtle.left(90)
    turtle.fd(size/2)
    turtle.rt(90)
    turtle.fd(int(size/3))
    turtle.back(int(size/3))
    turtle.lt(90)
    turtle.fd(size/2)
    turtle.rt(90)
    turtle.fd(int(size/3))
    
def S(size):
    turtle.rt(90)
    for i in range(0,5):
        if i<3:
            turtle.fd(size/2)
            turtle.lt(90)
            if i==2:
                turtle.rt(90)
        else:
            turtle.right(90)
            turtle.fd(size/2)

def O(size):
    turtle.circle(size)
    
def move(x,y):
    turtle.up()
    turtle.setposition(0,0)
    turtle.setheading(90)
    turtle.rt(90)
    turtle.fd(x)
    turtle.lt(90)
    turtle.fd(y)
    turtle.pendown()
    
def mov(x,y):
    turtle.up()
    turtle.setposition(0,0)
    turtle.setheading(90)
    turtle.lt(90)
    turtle.fd(x)
    turtle.rt(90)
    turtle.fd(y)
    turtle.pendown()  
    
def curve():
    for i in range(200):
        turtle.right(1)
        turtle.forward(1)

turtle.speed(0)          
move(550,200)
turtle.setposition(550,200)
turtle.setheading(0)
turtle.color("red","pink")
turtle.begin_fill()
turtle.left(140)
turtle.forward(111.65)
curve()

turtle.left(120)
curve()
turtle.forward(111.65)
turtle.end_fill()
turtle.hideturtle()
          
mov(550,-380)
turtle.setposition(-550,-380)
turtle.setheading(0)
turtle.color("red","pink")
turtle.begin_fill()
turtle.left(140)
turtle.forward(111.65)
curve()

turtle.left(120)
curve()
turtle.forward(111.65)
turtle.end_fill()
turtle.hideturtle()
        
turtle.speed(0)          
move(0,-170)
turtle.setposition(0,-170)
turtle.setheading(0)
turtle.color("red","pink")
turtle.begin_fill()
turtle.left(140)
turtle.forward(111.65)
curve()

turtle.left(120)
curve()
turtle.forward(111.65)
turtle.end_fill()
turtle.hideturtle()  

turtle.speed(5)  
turtle.pencolor("pink")  
mov(305,30)
H(1)
mov(265,30)
A(65)
mov(220,30)
P(1)
mov(185,30)
P(1)
mov(137,30)
Y(1)    


mov(85,30)
M(110)
move(30,52)
turtle.circle(27)
move(40,30)
T(60)
move(60,30)
H(1)
move(100,30)
E(65)
move(140,30)
R(1)
move(170,90)
turtle.fd(10)
move(180,30)
S(63)

move(255,30)
D(1)
move(295,30)
A(65)
move(342,30)
Y(1)


scr.exitonclick()
```
